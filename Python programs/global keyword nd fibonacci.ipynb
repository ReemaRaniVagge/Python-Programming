```python
#fibonacci sequence
#0+1=1,1+1=2,1+2=3,2+3=5,3+5=8 -> 0,1,1,2,3,5,8,13,21...

def fib(n):
    a=0
    b=1

    print(a)
    print(b)
    for i in range(2,n):   #0,1 is mentioned so we r starting from 2
        c=a+b
        a=b
        b=c
        print(c)
fib(9)
```

```python
#if u want to print only one number in series then below code

def fib(n):
    a=0
    b=1
    if n==1:
        print(a)
    else:

        print(a)
        print(b)
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            print(c)
fib(1)
```

```python
#factorial number


def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f        
x=4
result=fact(x)
print (result)
```
